```python
import tkinter
from tkinter import *

window = Tk() 
window.title('Simple Calculator')
window.geometry('300x500')
window.configure(bg="darkgrey")

equation = ""

def show(value):
    global equation
    equation += str(value)
    label1.config(text=equation)

def clear():
    global equation
    equation = ""
    label1.config(text=equation)
    
def calculate():
    global equation 
    try:
        result = str(eval(equation))  
        label1.config(text=result)
        equation = result  
    except Exception as e:
        label1.config(text="Error")
        equation = ""

label1 = Label(window, width=20, height=1, text="", font=("arial", 30))
label1.pack()

Button(window, text="AC", width=2, font=("arial", 30, "bold"), bd=1, bg="blue", command=clear).place(x=5, y=55)
Button(window, text="%", width=2, font=("arial", 30, "bold"), bd=1, bg="grey", command=lambda: show("%")).place(x=75, y=55)
Button(window, text="(", width=2, font=("arial", 30, "bold"), bd=1, bg="grey", command=lambda: show("(")).place(x=140, y=55)
Button(window, text=")", width=2, font=("arial", 30, "bold"), bd=1, bg="grey", command=lambda: show(")")).place(x=200, y=55)

Button(window, text="9", width=2, font=("arial", 30, "bold"), bd=1, bg="white", command=lambda: show("9")).place(x=5, y=140)
Button(window, text="8", width=2, font=("arial", 30, "bold"), bd=1, bg="white", command=lambda: show("8")).place(x=75, y=140)
Button(window, text="7", width=2, font=("arial", 30, "bold"), bd=1, bg="white", command=lambda: show("7")).place(x=140, y=140)
Button(window, text="+", width=2, font=("arial", 30, "bold"), bd=1, bg="grey", command=lambda: show("+")).place(x=200, y=140)

Button(window, text="6", width=2, font=("arial", 30, "bold"), bd=1, bg="white", command=lambda: show("6")).place(x=5, y=220)
Button(window, text="5", width=2, font=("arial", 30, "bold"), bd=1, bg="white", command=lambda: show("5")).place(x=75, y=220)
Button(window, text="4", width=2, font=("arial", 30, "bold"), bd=1, bg="white", command=lambda: show("4")).place(x=140, y=220)
Button(window, text="-", width=2, font=("arial", 30, "bold"), bd=1, bg="grey", command=lambda: show("-")).place(x=200, y=220)

Button(window, text="3", width=2, font=("arial", 30, "bold"), bd=1, bg="white", command=lambda: show("3")).place(x=5, y=300)
Button(window, text="2", width=2, font=("arial", 30, "bold"), bd=1, bg="white", command=lambda: show("2")).place(x=75, y=300)
Button(window, text="1", width=2, font=("arial", 30, "bold"), bd=1, bg="white", command=lambda: show("1")).place(x=140, y=300)
Button(window, text="*", width=2, font=("arial", 30, "bold"), bd=1, bg="grey", command=lambda: show("*")).place(x=200, y=300)

Button(window, text=".", width=2, font=("arial", 30, "bold"), bd=1, bg="white", command=lambda: show(".")).place(x=5, y=380)
Button(window, text="0", width=2, font=("arial", 30, "bold"), bd=1, bg="white", command=lambda: show("0")).place(x=75, y=380)
Button(window, text="=", width=2, font=("arial", 30, "bold"), bd=1, bg="orange", command=calculate).place(x=140, y=380)
Button(window, text="/", width=2, font=("arial", 30, "bold"), bd=1, bg="grey", command=lambda: show("/")).place(x=200, y=380)

window.mainloop()
```
